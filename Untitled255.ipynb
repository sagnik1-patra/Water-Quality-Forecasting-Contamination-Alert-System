```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# ============================================================
#  Paths
# ============================================================
base_dir = r"C:\Users\NXTWAVE\Downloads\Water Quality Forecasting & Contamination Alert System"
data_path = os.path.join(base_dir, "archive", "water_potability.csv")
visual_dir = os.path.join(base_dir, "visuals")
os.makedirs(visual_dir, exist_ok=True)

# ============================================================
#  Load dataset
# ============================================================
print("[INFO] Loading dataset...")
df = pd.read_csv(data_path)
df = df.fillna(df.mean(numeric_only=True))

X = df.drop("Potability", axis=1)
y = df["Potability"]

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, os.path.join(base_dir, "ais_csa_hydroguard_scaler.pkl"))

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# ============================================================
#  LSTMGRU Model Builder
# ============================================================
def build_model(lr, lstm_units, gru_units, dropout_rate):
    model = Sequential([
        LSTM(lstm_units, return_sequences=True, input_shape=(1, X_train.shape[2])),
        GRU(gru_units),
        Dropout(dropout_rate),
        Dense(1, activation="sigmoid")
    ])
    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return model

# ============================================================
#  Hybrid AIS + CSA Optimizer
# AIS = Immune-based mutation
# CSA = Crow Search Algorithm (follow the best solution)
# ============================================================

def hybrid_ais_csa(objective_func, bounds, n_agents=10, n_iter=10):
    dim = len(bounds)
    lb = np.array([b[0] for b in bounds])
    ub = np.array([b[1] for b in bounds])

    # Initialize population
    pop = np.random.rand(n_agents, dim)
    memory = pop.copy()

    best_global = None
    best_global_score = float("inf")

    for t in range(n_iter):
        awareness_prob = 0.2   # Crow awareness
        flight_length = 2

        for i in range(n_agents):
            if np.random.rand() > awareness_prob:
                # Follow memory of a random crow
                r = np.random.randint(n_agents)
                new_pos = pop[i] + flight_length * np.random.rand(dim) * (memory[r] - pop[i])
            else:
                # AIS mutation (exploration)
                new_pos = pop[i] + np.random.normal(0, 0.1, dim)

            new_pos = np.clip(new_pos, 0, 1)
            params = lb + new_pos * (ub - lb)
            score = objective_func(params)

            # Update memory if improved
            if score < best_global_score:
                best_global_score = score
                best_global = new_pos.copy()

            pop[i] = new_pos

        print(f"[Iter {t+1}/{n_iter}] Best Score = {best_global_score:.5f}")

    return lb + best_global * (ub - lb)

# ============================================================
#  Objective Function
# ============================================================
def objective(params):
    lr, lstm_units, gru_units, dropout = params
    lstm_units = int(lstm_units)
    gru_units = int(gru_units)

    model = build_model(lr, lstm_units, gru_units, dropout)
    model.fit(X_train, y_train, epochs=8, batch_size=32, verbose=0)

    _, acc = model.evaluate(X_test, y_test, verbose=0)
    return 1 - acc  # minimize

# ============================================================
#  Execute Optimization
# ============================================================
param_bounds = [
    (0.0005, 0.005),  # learning rate
    (16, 128),        # LSTM units
    (16, 128),        # GRU units
    (0.1, 0.5)        # dropout
]

best_params = hybrid_ais_csa(
    objective_func=objective,
    bounds=param_bounds,
    n_agents=6,
    n_iter=5
)

best_lr, best_lstm, best_gru, best_dropout = best_params
best_lstm = int(best_lstm)
best_gru = int(best_gru)

# Save config
config = {
    "optimizer": "Hybrid AIS + CSA",
    "best_params": {
        "learning_rate": float(best_lr),
        "lstm_units": best_lstm,
        "gru_units": best_gru,
        "dropout": float(best_dropout)
    }
}
with open(os.path.join(base_dir, "ais_csa_hydroguard_config.yaml"), "w") as f:
    yaml.dump(config, f)

# ============================================================
#  Final Model Training
# ============================================================
final_model = build_model(best_lr, best_lstm, best_gru, best_dropout)

history = final_model.fit(
    X_train, y_train,
    epochs=25,
    batch_size=32,
    validation_data=(X_test, y_test),
    verbose=1
)

final_model.save(os.path.join(base_dir, "ais_csa_hydroguard_model.h5"))

# ============================================================
#  Predictions & Evaluation
# ============================================================
y_pred = final_model.predict(X_test)
y_pred_class = (y_pred > 0.5).astype(int)

mae = mean_absolute_error(y_test, y_pred_class)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_class))
r2 = r2_score(y_test, y_pred_class)

metrics = {"MAE": mae, "RMSE": rmse, "R2": r2}

with open(os.path.join(base_dir, "ais_csa_hydroguard_prediction.json"), "w") as f:
    json.dump(metrics, f, indent=4)

# Save result CSV
pd.DataFrame({
    "Actual": y_test,
    "Predicted": y_pred_class.flatten()
}).to_csv(os.path.join(base_dir, "ais_csa_hydroguard_result.csv"), index=False)

# ============================================================
#  Visualizations
# ============================================================
plt.figure()
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("AIS+CSA Accuracy")
plt.savefig(os.path.join(visual_dir, "ais_csa_hydroguard_accuracy.png"))

plt.figure()
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("AIS+CSA Loss")
plt.savefig(os.path.join(visual_dir, "ais_csa_hydroguard_loss.png"))

plt.figure()
sns.heatmap(
    pd.DataFrame({"Actual": y_test, "Predicted": y_pred_class.flatten()}).corr(),
    annot=True, cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.savefig(os.path.join(visual_dir, "ais_csa_hydroguard_heatmap.png"))

plt.figure()
plt.scatter(y_test, y_pred_class, alpha=0.5)
plt.title("Actual Vs Predicted")
plt.savefig(os.path.join(visual_dir, "ais_csa_hydroguard_comparison.png"))

print(f"[INFO] All AIS+CSA HydroGuard files saved successfully in:\n{base_dir}")
```
