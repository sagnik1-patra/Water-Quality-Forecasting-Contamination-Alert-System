```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import random

# ============================================================
# 📂 Paths
# ============================================================
base_dir = r"C:\Users\NXTWAVE\Downloads\Water Quality Forecasting & Contamination Alert System"
data_path = os.path.join(base_dir, "archive", "water_potability.csv")
visual_dir = os.path.join(base_dir, "visuals")
os.makedirs(visual_dir, exist_ok=True)

# ============================================================
# 📊 Load and preprocess dataset
# ============================================================
print("[INFO] Loading dataset...")
df = pd.read_csv(data_path)
df = df.fillna(df.mean(numeric_only=True))
print(f"[INFO] Dataset shape: {df.shape}")

X = df.drop('Potability', axis=1)
y = df['Potability']

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, os.path.join(base_dir, "hybrid_hydroguard_scaler.pkl"))

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# ============================================================
# 🧬 Define LSTM–GRU Model Builder
# ============================================================
def build_model(lr, lstm_units, gru_units, dropout_rate):
    model = Sequential([
        LSTM(lstm_units, return_sequences=True, input_shape=(1, X_train.shape[2])),
        GRU(gru_units),
        Dropout(dropout_rate),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=Adam(learning_rate=lr),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

# ============================================================
# 🦠 Hybrid AIS + PSO Optimizer (Simplified Implementation)
# ============================================================

def hybrid_ais_pso(objective_func, bounds, n_agents=10, n_iter=10):
    dim = len(bounds)
    lb = np.array([b[0] for b in bounds])
    ub = np.array([b[1] for b in bounds])

    # Initialize population
    pop = np.random.rand(n_agents, dim)
    velocities = np.random.randn(n_agents, dim) * 0.1
    personal_best = pop.copy()
    personal_best_scores = np.array([float('inf')] * n_agents)
    global_best = None
    global_best_score = float('inf')

    for t in range(n_iter):
        for i in range(n_agents):
            params = lb + pop[i] * (ub - lb)
            score = objective_func(params)
            if score < personal_best_scores[i]:
                personal_best_scores[i] = score
                personal_best[i] = pop[i].copy()
            if score < global_best_score:
                global_best_score = score
                global_best = pop[i].copy()

        # PSO velocity & position updates
        w = 0.7
        c1, c2 = 1.5, 1.5
        for i in range(n_agents):
            r1, r2 = np.random.rand(dim), np.random.rand(dim)
            velocities[i] = w * velocities[i] + c1 * r1 * (personal_best[i] - pop[i]) + c2 * r2 * (global_best - pop[i])
            pop[i] += velocities[i]
            pop[i] = np.clip(pop[i], 0, 1)

        # AIS-inspired mutation and cloning
        clone_factor = 2
        clones = []
        for i in range(n_agents):
            for _ in range(clone_factor):
                mutant = pop[i] + np.random.normal(0, 0.05, dim)
                clones.append(np.clip(mutant, 0, 1))
        clones = np.array(clones)

        # Evaluate clones
        for clone in clones:
            params = lb + clone * (ub - lb)
            score = objective_func(params)
            if score < global_best_score:
                global_best_score = score
                global_best = clone.copy()

        print(f"[Iter {t+1}/{n_iter}] Best Score = {global_best_score:.5f}")

    return lb + global_best * (ub - lb)

# ============================================================
# 🎯 Objective Function
# ============================================================
def objective(params):
    lr, lstm_units, gru_units, dropout = params
    lstm_units, gru_units = int(lstm_units), int(gru_units)
    model = build_model(lr, lstm_units, gru_units, dropout)
    model.fit(X_train, y_train, epochs=8, batch_size=32, verbose=0)
    loss, acc = model.evaluate(X_test, y_test, verbose=0)
    return 1 - acc

# ============================================================
# 🚀 Run Optimization
# ============================================================
param_bounds = [
    (0.0005, 0.005),   # Learning Rate
    (16, 128),         # LSTM Units
    (16, 128),         # GRU Units
    (0.1, 0.5)         # Dropout
]

best_params = hybrid_ais_pso(objective, param_bounds, n_agents=6, n_iter=5)
best_lr, best_lstm, best_gru, best_dropout = best_params
best_lstm, best_gru = int(best_lstm), int(best_gru)

config = {
    "optimizer": "Hybrid AIS + PSO",
    "best_params": {
        "learning_rate": float(best_lr),
        "lstm_units": best_lstm,
        "gru_units": best_gru,
        "dropout": float(best_dropout)
    }
}
with open(os.path.join(base_dir, "hybrid_hydroguard_config.yaml"), "w") as f:
    yaml.dump(config, f)

# ============================================================
# 🧠 Train Final Model
# ============================================================
final_model = build_model(best_lr, best_lstm, best_gru, best_dropout)
history = final_model.fit(X_train, y_train, epochs=25, batch_size=32,
                          validation_data=(X_test, y_test), verbose=1)

final_model.save(os.path.join(base_dir, "hybrid_hydroguard_model.h5"))

# ============================================================
# 📈 Evaluate and Save Results
# ============================================================
y_pred = final_model.predict(X_test)
y_pred_class = (y_pred > 0.5).astype(int)

mae = mean_absolute_error(y_test, y_pred_class)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_class))
r2 = r2_score(y_test, y_pred_class)

metrics = {'MAE': mae, 'RMSE': rmse, 'R2': r2}
print(f"[RESULT] MAE={mae:.4f}, RMSE={rmse:.4f}, R²={r2:.4f}")

res_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_class.flatten()})
res_df.to_csv(os.path.join(base_dir, "hybrid_hydroguard_result.csv"), index=False)

with open(os.path.join(base_dir, "hybrid_hydroguard_prediction.json"), "w") as f:
    json.dump(metrics, f, indent=4)

# ============================================================
# 📊 Visualization
# ============================================================
plt.figure()
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title("Model Accuracy")
plt.legend()
plt.savefig(os.path.join(visual_dir, "hybrid_hydroguard_accuracy.png"))

plt.figure()
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("Model Loss")
plt.legend()
plt.savefig(os.path.join(visual_dir, "hybrid_hydroguard_loss.png"))

plt.figure()
sns.heatmap(pd.DataFrame({"Actual": y_test, "Predicted": y_pred_class.flatten()}).corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.savefig(os.path.join(visual_dir, "hybrid_hydroguard_heatmap.png"))

plt.figure()
plt.scatter(y_test, y_pred_class, alpha=0.5)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.savefig(os.path.join(visual_dir, "hybrid_hydroguard_comparison.png"))

print(f"[INFO] ✅ All outputs saved in: {base_dir}")
```
